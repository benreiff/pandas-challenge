```python
# Setup for worksheet
import pandas as pd
import os
import numpy as np

# Import CSV files from their location and give them a name
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read converted CSV files and store in Pandas dataframe
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Merge the two original CSV files into one dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
school_data_complete["type"].value_counts()
```

```python
# Calculate the district summary with:
    #total schools
total_schools = len(school_data_complete["school_name"].unique())
  
    #total students
total_students = len(school_data_complete["Student ID"].value_counts())

    #total budget
total_budget = school_data["budget"].sum()
    
    #average read and math scores
district_math_average = student_data["math_score"].mean()

district_reading_average = student_data["reading_score"].mean()   
   
    #overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

combined_math_reading_score = (district_math_average + district_reading_average)/2

    #percentage of students passing math (70+)
passing_math_students = len(student_data.loc[student_data["math_score"] >= 70])
passing_math_percentage = (passing_math_students/total_students)*100

    #percentage of students passing reading (70+)
passing_reading_students = len(student_data.loc[student_data["reading_score"] >= 70])
passing_reading_percentage = (passing_reading_students/total_students)*100


    #give the displayed data cleaner formatting
district_summary = pd.DataFrame({"Total Schools": [total_schools],
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "District Math Average": [district_math_average],
                                 "District Reading Average": [district_reading_average],
                                 "Overall Passing Rate": [combined_math_reading_score],
                                 "% Passing Math": [passing_math_percentage],
                                 "% Passing Reading": [passing_reading_percentage]})

district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)


district_summary
```

```python
list_of_schools = school_data_complete.groupby(['school_name', 'type'])
list_of_schools.count()
```

```python
school_size = list_of_schools["size"].unique()
school_size
```

```python
school_budget = list_of_schools["budget"].unique()
school_budget
```

```python
# Calculate how much of the budget is spent per student
budget_per_student = (school_budget/school_size)
budget_per_student
```

```python
# Calculate the math average per school
# Add all math scores and divide by the total number of students per school
math_score_by_school = list_of_schools["math_score"].sum()
#math_score_by_school


math_average_by_school = (math_score_by_school/school_size)
math_average_by_school
```

```python
# Calculate the reading average per school
# Add all reading scores and divide by the total number of students per school
reading_score_by_school = list_of_schools["reading_score"].sum()
#reading_score_by_school


reading_average_by_school = (reading_score_by_school/school_size)
reading_average_by_school
```

```python
 # Overall Passing Rate - combined_average of math and reading score
math_reading_score = (math_average_by_school+reading_average_by_school)/2
math_reading_score
```

```python
# Calculate percentage of students per school passing math

#passing_math_percentage = list_of_schools("math_score") >= 70

#passing_math_percentage

passing_math_percentage = school_data_complete.loc[school_data_complete["math_score"] > 70]
passing_math_percentage
```

```python
school_summary_table = pd.DataFrame({"Total Students": school_size,
                                     "Total School Budget": school_budget,
                                     "Per Student Budget": budget_per_student,
                                     "Average Math Score": math_average_by_school,
                                     "Average Reading Score": reading_average_by_school,
                                     #"% Passing Math",
                                     #"% Passing Reading": 
                                     "Overall Passing Rate": math_reading_score})

school_summary_table
```

```python
# Create a table highlighting the Top 5 performing schools

top_five_schools = school_summary_table.sort_values("Overall Passing Rate", ascending=False)

top_five_schools.head()
```
